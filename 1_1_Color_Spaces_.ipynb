## **`Exercise #1 Color spaces`**





1. Take a color image and visualize it. Remember: The image willbe read in BGR (Blue-Green-Red), because of OpenCVdefaults. OpenCV imshow() function assumes that the image shown is in BGR color-space. Which means, you will always need to convert back to RGB to see what we want.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = cv2.imread('/content/flowers.jpg')  # image will be read in BGR

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # conversion BGR -> RGB

plt.imshow(img_rgb)
plt.show(plt.xticks([]), plt.yticks([]))
```


---
2. Resize the image using three different scales (e.g. 0.3, 0.5, 0.7), visualize each rescaled image and for each compute the total number of pixels

```python
scale1 = 0.3

img_copy1 = img_rgb.copy() # create a copy of the image in order to not modify the original image

width1 = int(img_copy1.shape[1]*scale1)
height1 =int(img_copy1.shape[0]*scale1)
dimension1=(width1, height1)
img1_resized = cv2.resize(img_copy1, dimension1, interpolation=cv2.INTER_AREA)

# visualize the image 1 resized and show its details
plt.imshow(img1_resized)
plt.show()

print("[Details]")
print("Image resized using a scale of:", scale1)
print("\nwidth={}, height={}".format(width1, height1))
print("\nThe total number of pixels is:", width1*height1)
```

```python
scale2 = 0.5

img_copy2 = img_rgb.copy()

width2 = int(img_copy2.shape[1]*scale2)
height2 = int(img_copy2.shape[0]*scale2)
dimension2=(width2, height2)
img2_resized = cv2.resize(img_copy2, dimension2, interpolation=cv2.INTER_AREA)


# visualize the image 2 resized and show its details
plt.imshow(img2_resized)
plt.show()

print("[Details]")
print("Image resized using a scale of:", scale2)
print("\nwidth={}, height={}".format(width2, height2))
print("\nThe total number of pixels is:", width2*height2)
```

```python
scale3 = 0.7

img_copy3 = img_rgb.copy()

width3 = int(img_copy3.shape[1]*scale3)
height3 =int(img_copy3.shape[0]*scale3)
dimension3=(width3, height3)
img3_resized = cv2.resize(img_copy3, dimension3, interpolation=cv2.INTER_AREA)

# visualize the image 3 resized and show its details
plt.imshow(img3_resized)
plt.show()

print("[Details]")
print("Image resized using a scale of:", scale3)
print("\nwidth={}, height={}".format(width3, height3))
print("\nThe total number of pixels is:", width3*height3)
```

3. Visualize the image using the color spaces: RGB, HSV and Lab.





```python
# RGB color space, obtained in the first point of this exercise
plt.imshow(img_rgb)
plt.show(plt.xticks([]), plt.yticks([]))
plt.show()
```

```python
# RGB -> HSV
img_hsv_copy = img_rgb.copy()
img_hsv = cv2.cvtColor(img_hsv_copy, cv2.COLOR_BGR2HSV)

plt.imshow(img_hsv)
plt.show(plt.xticks([]), plt.yticks([]))
plt.show()
```

```python
# RGB -> Lab
img_lab_copy = img_rgb.copy()
img_lab = cv2.cvtColor(img_lab_copy, cv2.COLOR_BGR2Lab)

plt.imshow(img_lab)
plt.show(plt.xticks([]), plt.yticks([]))
plt.show()
```

```python
# RGB -> HLS

hls_copy = img_rgb.copy()

img_hls = cv2.cvtColor(hls_copy, cv2.COLOR_BGR2HLS)
plt.imshow(img_hls)
plt.show(plt.xticks([]), plt.yticks([]))
plt.show()
```
