# **Exercise #2 - Drawing**

1. Take a color image and show it 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_bgr = mpimg.imread('burano.png')
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

imgplot = plt.imshow(img)
plt.show() 
```



---


2. Put a grid 3x3 over the image by drawing vertical and horizontal lines in green color



```python
img_copy = img.copy()

# image: width = 322, height = 571
# each rectangle of the grid will be 190x107

# vertical lines
cv2.line(img_copy, (190, 0), (190, 322), (0, 255, 0), 5)
cv2.line(img_copy, (380, 0), (380, 322), (0, 255, 0), 5)

# horizontal lines
cv2.line(img_copy, (0, 107), (571, 107), (0, 255, 0), 5)
cv2.line(img_copy, (0, 215), (571, 215), (0, 255, 0), 5)

plt.imshow(img_copy)
plt.show()
```

3. Consider the central pixel of your image and draw four arrows of different colors. Each arrow starts in a corner of the image and ends in the central pixel

```python
img_copy = img.copy()

c_width = int(img_copy.shape[1] / 2)
c_height = int(img_copy.shape[0] / 2)

center = (c_width , c_height)

cv2.arrowedLine(img_copy, (0, 0), center, (0, 0, 255), 5)
cv2.arrowedLine(img_copy, (571, 0), center, (255, 0, 0), 5)
cv2.arrowedLine(img_copy, (0, 322), center, (0, 255, 0), 5)
cv2.arrowedLine(img_copy, (571, 322), center, (255, 255, 0), 5)

plt.imshow(img_copy)
plt.show()
```
