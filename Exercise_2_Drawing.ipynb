# **Exercise #2 - Drawing**

1. Take a color image and show it 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('walloflove.jpg')

plt.imshow(img)
plt.show(plt.xticks([]), plt.yticks([])) 
```



---


2. Put a grid 3x3 over the image by drawing vertical and horizontal lines in green color



```python
img_copy = img.copy()

# image: width = 1024 , height = 670
# each rectangle of the grid will be 341x223

# vertical lines
cv2.line(img_copy, (341, 0), (341, 670), (0, 255, 0), 5)
cv2.line(img_copy, (682, 0), (682, 670), (0, 255, 0), 5)

# horizontal lines
cv2.line(img_copy, (0, 223), (1024, 223), (0, 255, 0), 5)
cv2.line(img_copy, (0, 446), (1024, 446), (0, 255, 0), 5)

plt.imshow(img_copy)
plt.show()
```



---

3. Consider the central pixel of your image and draw four arrows of different colors. Each arrow starts in a corner of the image and ends in the central pixel

```python
img_copy = img.copy()

c_width = int(img_copy.shape[1] / 2)
c_height = int(img_copy.shape[0] / 2)

center = (c_width , c_height)

cv2.arrowedLine(img_copy, (0, 0), center, (0, 255, 0), 5)
cv2.arrowedLine(img_copy, (1024, 0), center, (255, 0, 0), 5)
cv2.arrowedLine(img_copy, (0, 670), center, (0, 255, 255), 5)
cv2.arrowedLine(img_copy, (1024, 670), center, (255, 255, 0), 5)

plt.imshow(img_copy)
plt.show()
```

4. Create an empty black image and put your name on it using the font and the color your prefer


```python
w = 800
h = 800
size = (h, w, 3)

black_image = np.zeros(size, dtype=np.uint8)
cv2.putText(black_image, text='Vito Vicenti', org=(70, 350), fontFace=cv2.FONT_HERSHEY_COMPLEX , fontScale=3, color=(0, 255, 255), thickness=5)

plt.imshow(black_image)
plt.show()
```
